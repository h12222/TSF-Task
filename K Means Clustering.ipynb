# Taks 2:- IRIS dataset, predict the optimum number of clusters and represent it visually.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets  import load_iris
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
iris
```

```python
iris=load_iris()
x=pd.DataFrame(iris.data)
y=pd.DataFrame(iris.target)
y
```

```python
x.columns=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
y.columns=['Target']
x
```

```python
colormap=np.array(['red','blue','black'])
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.scatter(x.Sepal_Length,x.Sepal_Width,c=colormap[y.Target],s=40)
plt.title("Sepal")
plt.subplot(1,2,2)
plt.scatter(x.Petal_Length,x.Petal_Width,c=colormap[y.Target],s=40)
plt.title("Petal")
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3)
km.fit(x)
km.labels_
```

```python
colormap=np.array(['red','blue','black'])
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.scatter(x.Petal_Length,x.Petal_Width,c=colormap[y.Target],s=40)
plt.title("Real")
plt.subplot(1,2,2)
plt.scatter(x.Petal_Length,x.Petal_Width,c=colormap[km.labels_],s=40)
plt.title("kmeans")
```

```python
colormap=np.array(['red','blue','black'])
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.scatter(x.Sepal_Length,x.Sepal_Width,c=colormap[y.Target],s=40)
plt.title("Real")
plt.subplot(1,2,2)
plt.scatter(x.Sepal_Length,x.Sepal_Width,c=colormap[km.labels_],s=40)
plt.title("kmeans")
```

```python
print("Accuracy of Kmeans",accuracy_score(y,km.labels_))
```
